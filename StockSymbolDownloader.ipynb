```python
import json

def build_sec_query(tickers):
    if isinstance(tickers, list):
        tickers_query = " OR ".join(tickers)
        query_string = f'formType:"10-K" AND (ticker:{tickers_query})'
    else:
        query_string = f'formType:"10-K" AND ticker:{tickers}'

    query = {
        "query": query_string,
        "from": "0",
        "size": "50"
    }

    # Properly format JSON output
    return json.dumps(query, indent=4)


test_output = ['AAPL', 'MSFT', 'NVDA', 'AMZN', 'GOOGL', 'META', 'TSLA', 'BRK-B', 'AVGO', 'WMT','JPM', 'LLY', 'V', 'SPY', 'MA', 'UNH', 'XOM', 'ORCL', 'COST', 'NFLX', 'HD', 'PG', 'NVO']
generated_query = json.loads(build_sec_query(test_output))
assert query == expected_query, "Generated query does not match expected query"

print("Assertion passed: Generated query matches expected query")
query_result = build_sec_query(test_output)
print(query_result)
```

```python
first_entry_urls = []
for item in df.loc[0, 'documentFormatFiles']:
    if 'documentUrl' in item:
        first_entry_urls.append(item['documentUrl'])

print(first_entry_urls)
```

```python
def get_first_document_urls(df):
    ticker_urls = {}
    for ticker in df['ticker'].unique():
        try:
            first_row = df[df['ticker'] == ticker].iloc[0]
            first_url = first_row['documentFormatFiles'][0]['documentUrl']
            ticker_urls[ticker] = first_url
        except (IndexError, KeyError, TypeError):
            ticker_urls[ticker] = None
    return ticker_urls
urls=get_first_document_urls(df)
print(urls)
```

```python
df_urls = pd.DataFrame(list(urls.items()), columns=['Ticker', 'First Document URL'])
pd.set_option('display.max_colwidth', None)
print(df_urls.head())
```

```python
from sec_api import PdfGeneratorApi
from google.colab import drive
import os

# Mount Google Drive
drive.mount('/content/drive')

# Assuming 'SEC_API_KEY' is already defined
pdfGeneratorApi = PdfGeneratorApi(api_key=SEC_API_KEY)

# Create the '10k_filings' folder if it doesn't exist
folder_path = '/content/drive/My Drive/10k_filings'
os.makedirs(folder_path, exist_ok=True)

for url in df_urls['First Document URL'].unique():
    try:
        ticker = url.split('/')[-1].split('.')[0]

        file_path = os.path.join(folder_path, f"{ticker}_10K.pdf")

        # Check if file already exists
        if os.path.exists(file_path):
            print(f"File for {ticker} already exists. Skipping...")
            continue

        pdf_10K_filing = pdfGeneratorApi.get_pdf(url)

        with open(file_path, 'wb') as f:
            f.write(pdf_10K_filing)
        print(f"Saved 10-K filing for {ticker} to Google Drive.")

    except Exception as e:
        print(f"Error processing URL {url}: {e}")
```

```python
from langchain.document_loaders import DirectoryLoader, UnstructuredPDFLoader
from langchain.text_splitter import RecursiveCharacterTextSplitter
import nltk
nltk.download('punkt_tab')
nltk.download('averaged_perceptron_tagger_eng')
# Define the directory path
directory = '/content/drive/My Drive/10k_filings'

# Initialize the loaders and text splitter
loader = DirectoryLoader(directory, glob="**/*.pdf", loader_cls=UnstructuredPDFLoader)
text_splitter = RecursiveCharacterTextSplitter(chunk_size=1000, chunk_overlap=100)

# Load and split the documents
documents = loader.load()
docs = text_splitter.split_documents(documents)
```
